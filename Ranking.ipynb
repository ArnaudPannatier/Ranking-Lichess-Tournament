```python
#tournaments_ids = ["bsWB13Hw", "YcZ1jadO", "4msjnkop", "6HR4y64q", "Ui5I3XN0", "ZNViLhlA", "rpwTJ8ba"]
tournaments_ids = ids

request_url_games = "https://lichess.org/api/tournament/{}/games?evals=true&opening=true"
request_url = "https://lichess.org/api/tournament/{}/results"
```

```python
D_list = []
for i, t_id in enumerate(tournaments_ids):
    response = requests.get(request_url.format(t_id))
    tournaments_dicts = [json.loads(s) for s in response.text.split("\n") if len(s) > 0]
    for d in tournaments_dicts : d["Tournament"] = i + 1

    df = pd.DataFrame(tournaments_dicts)
    index_tour = str(df.Tournament.values[0])
    df.drop(columns=["rating", "performance", "Tournament"], inplace=True)
    df.rename(columns={"rank" : "rank"+index_tour}, inplace=True)
    
    D_list.append(df)
```

```python
def process_df(x,y):
    df = pd.merge(x, y, on = 'username', how = 'outer')
    name = "score"
    df[name] = df[name+"_x"].fillna(0) + df[name+"_y"].fillna(0)

    
    df.drop(columns=[name+"_x", name+"_y"], inplace=True)
    
    return df

final_df = reduce(lambda x,y: process_df(x,y), D_list).fillna(' ')
cols = final_df.columns.values.copy()
cols[:2] = ["username", "rank1"]
final_df[cols].sort_values(by="score", ascending=False).reset_index(drop=True)
```
